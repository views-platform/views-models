```python
model = pd.read_pickle('models/orange_pasta/artifacts/forecasting_model_20241108_162549.pkl')
```

```python
from ingester3.ViewsMonth import ViewsMonth
month_last = ViewsMonth.now().id - 2
partitioner_dict = {"train":(121, month_last),"predict":(month_last +1, month_last + 1 + 36)}
train_start, train_end = partitioner_dict["train"]
test_start, test_end = partitioner_dict["predict"]
```

```python
df = pd.read_pickle('models/orange_pasta/data/raw/forecasting_viewser_df.pkl')
level = df.index.names[1]
time = df.index.names[0]
independent_variables = [c for c in df.columns if c != 'ln_ged_sb_dep']

df_reset = df.reset_index(level=[1])
series = TimeSeries.from_group_dataframe(df_reset, group_cols=level,
                                                       value_cols=independent_variables + ['ln_ged_sb_dep'])
target_train = [s.slice(train_start, train_end + 1)['ln_ged_sb_dep']
                              for s in series]  # ts.slice is different from df.slice
past_cov_train = [series.slice(train_start, train_end + 1)[independent_variables]
                                for series in series]
past_cov = [s[independent_variables] for s in series]
```

```python
model_1 = model.models[1]
pred = model_1.predict(n=2,
                series=target_train,
                # darts automatically locates the time period of past_covariates
                past_covariates=past_cov,
                show_warnings=False)
```

```python
index_tuples, df_list = [], []
for p in pred:
    # For forecasting only need to keep predictions with last-month-with-data, i.e., diagonal prediction
    df_pred = p.pd_dataframe().loc[[test_start + 1]]
    l = p.static_covariates.iat[0, 0]
    index_tuples.extend([(month, l) for month in df_pred.index])
    df_list.append(df_pred.values)

df_preds = pd.DataFrame(
    data=np.concatenate(df_list),
    index=pd.MultiIndex.from_tuples(index_tuples, names=[time, level]),
    columns=[f"step_pred_combined"]
)
```

```python
model_2 = model.models[2]
pred_2 = model_2.predict(n=3,
                series=target_train,
                # darts automatically locates the time period of past_covariates
                past_covariates=past_cov,
                show_warnings=False)

index_tuples, df_list = [], []
for p in pred_2:
    # For forecasting only need to keep predictions with last-month-with-data, i.e., diagonal prediction
    df_pred = p.pd_dataframe().loc[[test_start + 2]]
    l = p.static_covariates.iat[0, 0]
    index_tuples.extend([(month, l) for month in df_pred.index])
    df_list.append(df_pred.values)

df_preds_2 = pd.DataFrame(
    data=np.concatenate(df_list),
    index=pd.MultiIndex.from_tuples(index_tuples, names=[time, level]),
    columns=[f"step_pred_combined"]
)
```
